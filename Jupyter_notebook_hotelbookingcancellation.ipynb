```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

df=pd.read_csv("hotel_bookings.csv")
df.head(10)
```

```python
df["reservation_status_date"]=pd.to_datetime(df["reservation_status_date"])
```

```python
df.describe(include="object")
```

```python
df.isnull().sum()
```

```python
df.drop(["company","agent"],axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
for col in df.describe(include="object").columns:
    print(col)
    print(df[col].unique())
    print("-"*50)
```

```python
cancelled_percentage=df["is_canceled"].value_counts(normalize=True)
print(cancelled_percentage)
plt.bar(["not canceled","canceled"],df["is_canceled"].value_counts(),edgecolor="k",width=0.7)
plt.title("Reservation status count")
plt.show()
```

```python
a=sb.countplot(data=df,x="hotel",hue="is_canceled",palette="viridis")
a.bar_label(a.containers[0])
a.bar_label(a.containers[1])
plt.title("Reservation status in different Hotels")
plt.xlabel("Hotel")
plt.ylabel("Number of Reservations")
plt.legend(title="Hotel Bookings", labels=["Not Canceled", "Canceled"])
plt.show()
```

```python
resort_hotel=df[df["hotel"]=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
#72% not canceled 27% canceled
```

```python
city_hotel=df[df["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
#58% not canceled 41% canceled
```

```python
# Group by reservation date and calculate mean ADR
resort_hotel = resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel   = city_hotel.groupby("reservation_status_date")[["adr"]].mean()

# Convert index to datetime
resort_hotel.index = pd.to_datetime(resort_hotel.index)
city_hotel.index   = pd.to_datetime(city_hotel.index)

# Sort by date (important for line plots)
resort_hotel = resort_hotel.sort_index()
city_hotel   = city_hotel.sort_index()

# Plot
plt.figure(figsize=(10,6))
plt.title("Average Daily Rate in City and Resort Hotels")
plt.plot(resort_hotel.index, resort_hotel["adr"], label="Resort Hotel")
plt.plot(city_hotel.index, city_hotel["adr"], label="City Hotel")
plt.xlabel("Date")
plt.ylabel("Average Daily Rate (ADR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df["month"]=df["reservation_status_date"].dt.month
a=sb.countplot(data=df,x="month",hue="is_canceled")
plt.legend(labels=["not canelled","canceled"])
plt.xlabel("Months")
plt.ylabel("number of reservations")
plt.title("Reservation status per months")
a.bar_label(a.containers[0])
a.bar_label(a.containers[1])
plt.show()
```

```python
sb.barplot(x="month",y="adr",data=df[df["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index())
plt.title("Average daily Rate per months")
plt.show()
```

```python
cancelled_data=df[df["is_canceled"]==1]
top_10_country=cancelled_data["country"].value_counts()[:10]
plt.pie(top_10_country,autopct="%.2f",labels=top_10_country.index)
plt.title("top 10 countries with reservation cancelled")
plt.show()


```
