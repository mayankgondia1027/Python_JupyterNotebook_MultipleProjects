```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

connection=mysql.connector.connect(host="localhost",
                                   user="root",
                                   password="1818",
                                   database="banking_case")
query="""select * from banking_case.customers"""
df=pd.read_sql(query,connection)
df.head()

```

```python
bins=[0,100000,300000,float("inf")]
labels=["low","med","High"]

df['Income Band']=pd.cut(df["Estimated Income"],bins=bins, labels=labels, right=False)
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind="bar")
```

```python
#examine the distribution of unique categories in categorical columns

categorical_cols=df[['Nationality', 'Occupation', 'Fee Structure',
       'Loyalty Classification','Amount of Credit Cards','Properties Owned',
       'Risk Weighting', 'BRId', 'GenderId', 'IAId',"Income Band"]]

for col in categorical_cols:
       print(f"Value Counts for '{col}':")
       display(df[col].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
for i, predictor in enumerate(['Nationality', 'Occupation', 'Fee Structure',
       'Loyalty Classification', 'Amount of Credit Cards', 'Properties Owned',
       "Risk Weighting", "BRId", "GenderId", "IAId", "Income Band"]):

    plt.figure(i)
    sb.countplot(data=df, x=predictor, hue="GenderId",palette="viridis")
    plt.show()
    
    #univerate analysis
    #one column at one time example without hue
    
    #bivarate analysis
    #two column at one time example with hue
```

```python
#Histplot of value counts for different occupations

for col in categorical_cols:
    if col=="Occupation":
        continue
    sb.histplot(df[col])
    plt.title("Histogram of occupation count")
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
    
    
```

```python
#numerical analysis

numerical_cols=["Estimated Income", "Superannuation Savings","Credit Card Balance",
                "Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts",
                "Foreign Currency Account","Business Lending"]

for i ,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sb.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()
```

```python
#heatmaps

numerical_cols=["Estimated Income", "Superannuation Savings","Credit Card Balance",
                "Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts",
                "Foreign Currency Account","Business Lending"]

correlation_matrix=df[numerical_cols].corr()
for i ,col in enumerate(numerical_cols):
    sb.heatmap(correlation_matrix,annot=True,cmap="crest",fmt=".2f")
    plt.title("Correlation matrix")
    plt.show()
```
